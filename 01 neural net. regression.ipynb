```python
def plot_predictions(train_data=x_train,
                     train_labels=y_train,
                     test_data=x_test,
                     test_lebals=y_test,
                     predictions=y_preds):
    """
    plots training data,test data and compares prediction to ground truth labels.
    """
    plt.figure(figsize=(10,7))
    plt.scatter(train_data,train_labels,c='b',label='Training data')
    plt.scatter(test_data,test_lebals,c='g',label='Testing data')
    plt.scatter(test_lebals,predictions,c='r',label='Prediction')
    plt.legend();
```

```python
def mse(y_true,y_preds):
    return tf.keras.losses.mse(y_true=y_true,y_pred=tf.squeeze(y_preds))

def mae(y_true,y_preds):
    return tf.keras.losses.mae(y_true=y_true,y_pred=tf.squeeze(y_preds))
```

```python
insurance=pd.read_csv('https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv')
insurance
```

```python
# try one-hot encode from pandas
insurance_one_hot=pd.get_dummies(insurance,dtype=int)
insurance_one_hot.head()
```

```python
x=insurance_one_hot.drop('charges',axis=1)
y=insurance_one_hot['charges']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
len(x),len(x_train),len(x_test)
```

```python
# building model

tf.random.set_seed(42)

insurance_model=tf.keras.Sequential([
    tf.keras.layers.Dense(50,activation='relu'),
    tf.keras.layers.Dense(50,activation='relu'),
    tf.keras.layers.Dense(1)
])

insurance_model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(0.01),
              metrics=['mae'])

history=insurance_model.fit(x_train,y_train,epochs=200)
```

```python
insurance_model.evaluate(x_test,y_test)
```

```python
y_pred_insurance=insurance_model.predict(x_test)
y_pred_insurance
```

## normalisation

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler

ct=make_column_transformer(
    (MinMaxScaler(),['age','bmi','children']),
    (OneHotEncoder(handle_unknown='ignore'),['sex','smoker','region'])
)
x=insurance.drop('charges',axis=1)
y=insurance['charges']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)

ct.fit(x_train)

x_train_normal=ct.transform(x_train)
x_test_normal=ct.transform(x_test)
```

```python
x_train_normal[0]
```

```python
x_test_normal[0]
```

```python
x_train.shape,x_train_normal.shape
```

```python
# building model

normal_model_insurance=tf.keras.Sequential([
    tf.keras.layers.Dense(50,activation='relu'),
    tf.keras.layers.Dense(50,activation='relu'),
    tf.keras.layers.Dense(1)
])

normal_model_insurance.compile(loss=tf.keras.losses.mae,
             optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
             metrics=['mae'])

normal_model_insurance.fit(x_train_normal,y_train,epochs=200)
```

```python
normal_model_insurance.evaluate(x_test_normal,y_test)
```

```python
y_preds_normal_insurance=normal_model_insurance.predict(x_test_normal)
y_preds_normal_insurance
```
